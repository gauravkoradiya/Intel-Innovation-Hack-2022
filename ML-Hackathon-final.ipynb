# <div style="padding:20px;color:white;margin:0;font-size:100%;text-align:center;display:fill;border-radius:10px;background-color:#251cab;background:linear-gradient(30deg, #3eb4f4, #251CAB,#211894,#251CAB, #3eb4f4)">Data Description</div>

The dataset consists of about 4.3 million water samples that have been collected across the globe with 22 water quality attributes that include the pH, Iron, Nitrate, Chloride, Zinc, Color, Temperature, and Turbidity levels. It also contains the source of the water (ground, river, lake, etc.) as well as the date and time the sample was collected. 

Below is the description of the data files included for the Hackathon:  
- <b>water_quality_train.csv</b>: the training set consisting of the following features:
    - `pH` - the pH level of the water sample.
    - `Iron` - the total concentration of iron in the water sample, measured in milligrams per liter (mg/L).
    - `Nitrate` - the total concentration of nitrate in the water sample, measured in milligrams per liter (mg/L).
    - `Chloride` - the total concentration of chloride in the water sample, measured in milligrams per liter (mg/L).
    - `Lead` - the total concentration of lead in the water sample, measured in milligrams per liter (mg/L).
    - `Zinc` - the total concentration of zinc in the water sample, measured in milligrams per liter (mg/L).
    - `Color` - the color of the water sample: colorless, near colorless, faint yellow, light yellow, and yellow.
    - `Turbidity` - the cloudiness of the water sample, measured in Nephelometric Turbidity Units (NTU).
    - `Fluoride` - the total concentration of fluoride in the water sample, measured in milligrams per liter (mg/L).
    - `Copper` - the total concentration of copper in the water sample, measured in milligrams per liter (mg/L).
    - `Odor` - the threshold odor number (TON) of the water sample.
    - `Sulfate` - the total concentration of sulfate in the water sample, measured in milligrams per liter (mg/L).
    - `Conductivity` - the ability of water to pass an electrical current, measured in micromhos per centimeter (Âµmhos/cm).
    - `Chlorine` - the total concentration of chlorine in the water sample, measured in milligrams per liter (mg/L).
    - `Manganese` - the total concentration of manganese in the water sample, measured in milligrams per liter (mg/L).
    - `Total Dissolved Solids` - the total concentration of dissolved substances in the water sample, made up of organic and inorganic materials, such as metals, minerals, salts, and ions, measured in milligrams per liter (mg/L).
    - `Source` - the source of the water sample: ground, lake, river, stream, aquifer, reservoir, well, spring.  
    - `Water Temperature` - the temperature of the water that was collected, measured in degrees Fahrenheit (&deg;F).
    - `Air Temperature` - the temperature of the air when the water was collected, measured in degrees Celcius (&deg;C).
    - `Month` - the month the water sample was collected.
    - `Day` - the day of the month the water sample was collected.
    - `Time of Day` - the hour the water sample was collected.
    - `Target` - the target variable indicating the quality of the water sample. 
- <b>water_quality_test.csv</b>: the test set. For each sample, you should predict whether the water is safe (`Target = 0`) or unsafe (`Target = 1`).

## <span style="color:#211894;font-size:100%;text-align:left">Evaluation Criteria</span>

Submissions in the Machine Learning Track will be evaluated based on the [$F_{1}$ score](https://en.wikipedia.org/wiki/F-score). The $F_{1}$ score represents the harmonic mean between the Precision and Recall of the positive class, indicated by a 1. It can be derived using the following calculation: 

<br>

$$F_1 = \frac{2 * Precision * Recall}{Precision + Recall} = \frac{2 * TP}{2 * TP + FP + FN}$$

```python
def plot_target_dist(target_col):
    
    """Function to display distribution of the target variable"""
    
    target=data[target_col].value_counts(normalize=True)
    target.rename(index={1:'UnSafe Water',0:'Safe Water'},inplace=True)
    fig=go.Figure()
    fig.add_trace(go.Pie(labels=target.index, values=target*100, hole=.45, 
                         text=target.index, sort=False, showlegend=False,
                         marker=dict(colors=color,line=dict(color=intel_pal,width=2.5)),
                         hovertemplate = "%{label}: <b>%{value:.2f}%</b><extra></extra>"))
    fig.update_layout(template=temp, title='Target Distribution',width=700,height=450,
                      uniformtext_minsize=15, uniformtext_mode='hide')
    fig.show() 
    
plot_target_dist(target_col='Target')
```
